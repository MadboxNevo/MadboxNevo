## **Scraping Data From Website + Pandas**

```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_largest_companies_in_the_United_States_by_revenue'

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup)
```

```python
soup.find('table')
```

```python
soup.find_all('table')
```

```python
soup.find_all('table')[1]
```

```python
soup.find('table', class_ = 'wikitable sortable')
```

```python
table = soup.find_all('table')[1]
```

```python
print(table)
```

```python
soup.find_all('th')
```

```python
world_titles = table.find_all('th')
```

```python
world_titles
```

**Start building table**

```python
# world_table_titles = [ title.text.strip() for title in world_titles]
# print(world_table_titles)
```

```python
#there's sth in table we dont want which is, secondary table-track back issues, issue @ world_titles = table.find_all('th') not soup.find_all
```

```python
world_table_titles = [ title.text.strip() for title in world_titles]
print(world_table_titles)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = world_table_titles)

df
```

**start pulling in data**

```python
table.find_all('tr')

```

```python
column_data = table.find_all('tr')
```

```python
for row in column_data[1:]:
  row_data = row.find_all('td')
  individual_row_data = [data.text.strip() for data in row_data]
  length = len(df)
  df.loc[length] = individual_row_data
```

```python
df
```

**Export as CSV**

```python
df.to_csv(r'', index = False) #fill in path to export as csv
```
