**Population Distribution By Continent**

```python

## Population Distribution by Continent
continent_population = df4.groupby('Continent')['2022 Population'].sum()
plt.figure(figsize=(20, 10))
plt.pie(continent_population, labels=continent_population.index, autopct='%1.2f%%', startangle=140)
plt.title('2022 Population Distribution by Continent')
plt.show()
```

```python
## Relationship between Population Density and Area
plt.figure(figsize=(10, 7))
sns.scatterplot(x='Area (km²)', y='Density (per km²)', hue='Continent', data=df4, palette='Set1', s=100)
plt.title('Population Density vs Area by Continent')
plt.xlabel('Area (km²)')
plt.ylabel('Density (per km²)')
plt.show()

```

```python
## Distribution of Growth Rates
plt.figure(figsize=(10, 7))
sns.histplot(df4['Growth Rate'], bins=10, kde=True)
plt.title('Distribution of Growth Rates')
plt.xlabel('Growth Rate')
plt.ylabel('Frequency')
plt.show()

```

```python
## World Population Percentage - Top 10 Countries
top_countries_by_population_percentage = df4.nlargest(10, 'World Population Percentage')
plt.figure(figsize=(14, 7))
sns.barplot(x='World Population Percentage', y='CCA3', data=top_countries_by_population_percentage, palette='coolwarm')
plt.title('Top Countries by World Population Percentage')
plt.xlabel('World Population Percentage')
plt.ylabel('Country')
plt.show()
```
