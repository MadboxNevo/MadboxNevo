# Data Cleaning in Pandas

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = df = pd.read_excel('/content/Customer Call List.xlsx')
df
```

```python
df = df.drop_duplicates()
df
```

```python
df1 = df.drop(columns = "Not_Useful_Column")
df1
```

```python
#df["Last_Name"] = df["Last_Name"].str.lstrip("...")
#df["Last_Name"] = df["Last_Name"].str.lstrip("/")
#df["Last_Name"] = df["Last_Name"].str.rstrip("_")
df1["Last_Name"] = df1["Last_Name"].str.strip("123._/")
df1
```

```python
df1["Phone_Number"] = df1["Phone_Number"].str.replace('[^a-zA-Z0-9]','-')

# df["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])

# df["Phone_Number"] = df["Phone_Number"].apply(lambda x: str(x))

# df["Phone_Number"] = df["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])

df1["Phone_Number"] = df1["Phone_Number"].str.replace('nan--','')

df1["Phone_Number"] = df1["Phone_Number"].str.replace('Na--','')

# df["Phone_Number"] = df["Phone_Number"].str.replace('/','-')

# df["Phone_Number"] = df["Phone_Number"].str.replace('|','-')

df1["Phone_Number"] = df1["Phone_Number"].str.replace('N-a','')

df1
```

```python
df1[["Street_Address", "State", "Zip_Code"]] = df1["Address"].str.split(',',2, expand=True)
df1
```

```python
df1["Do_Not_Contact"] = df1["Do_Not_Contact"].str.replace('Yes','Y')

df1["Do_Not_Contact"] = df1["Do_Not_Contact"].str.replace('No','N')
df1
```

```python
#df = df.replace('N/a','')
#df = df.replace('NaN','')


df1=df1.fillna('')
df1
```

```python
for x in df1.index:
    if df1.loc[x, "Do_Not_Contact"] == 'Y':
        df1.drop(x, inplace=True)

df1
```

```python
for x in df1.index:
    if df1.loc[x, "Phone_Number"] == '':
        df1.drop(x, inplace=True)

df1

#Another way to drop null values
#df = df.dropna(subset="Phone_Number"), inplace=True)
```

```python
df1 = df1.reset_index(drop=True)
df1
```
